```python
def count_tokens_and_chars(text):
    # Tokenize by splitting on whitespace
    tokens = text.split()
    token_count = len(tokens)
    char_count = len(text)
    return token_count, char_count

# Get input from the user
input_string = input("Enter some text: ")

# Calculate token and character counts
tokens, chars = count_tokens_and_chars(input_string)

# Display the results
print(f"Token Count: {tokens}")
print(f"Character Count: {chars}")
```
