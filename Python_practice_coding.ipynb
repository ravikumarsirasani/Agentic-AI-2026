## Reverse String

````python
## Reverse String
def reverse_string(s):
    return s[::-1]

print(reverse_string("agentic"))
````

## Palindrome

````python
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome("RAR"))
````

## Fibonacci Sequence

````python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci(10)
````

## VOWELS

````python
def count_vowels(s):
    return sum(1 for c in s.lower() if c in "aeiou")

print(count_vowels("Welcome to Generative AI...  Ravi"))
````

```
## Find duplicates in the list
```



````python
def find_duplicates(lst):
    seen, dup = set(), set()
    for x in lst:
        if x in seen:
            dup.add(x)
        else:
            seen.add(x)
    return list(dup)

print(find_duplicates([1,2,3,2,4,1,5]))
````
