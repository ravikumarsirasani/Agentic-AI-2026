## Reverse String

````python
## Reverse String
def reverse_string(s):
    return s[::-1]

print(reverse_string("agentic"))
````

## Palindrome

````python
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome("RAR"))
````

## Fibonacci Sequence

````python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b

fibonacci(10)
````

## VOWELS

````python
def count_vowels(s):
    return sum(1 for c in s.lower() if c in "aeiou")

print(count_vowels("Welcome to Generative AI...  Ravi"))
````

````python
def find_duplicates(lst):
    seen, dup = set(), set()
    for x in lst:
        if x in seen:
            dup.add(x)
        else:
            seen.add(x)
    return list(dup)

print(find_duplicates([1,2,3,2,4,1,5]))
````

````python
def primes(n):
    result = []
    for num in range(2, n+1):
        for i in range(2, int(num**0.5)+1):
            if num % i == 0:
                break
        else:
            result.append(num)
    return result

print(primes(20))
````

````python
def two_sum(nums, target):
    d = {}
    for i, n in enumerate(nums):
        if target - n in d:
            return [d[target-n], i]
        d[n] = i

print(two_sum([2,7,11,15], 9))
````
